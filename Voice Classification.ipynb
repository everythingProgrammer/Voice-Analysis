```python
import pandas as pd
```

```python
data = pd.read_csv('Voice-Dataset.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(data.corr(), annot=True, linewidth=1, linecolor='w')
```

```python
label = LabelEncoder()
```

```python
output_data = label.fit_transform(data['label'])
```

```python
output_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
training_data = data.iloc[:, :-1]
```

```python
training_data.shape
```

```python
data.shape
```

```python
X, x_test, Y, y_test = train_test_split(training_data, 
                                       output_data, 
                                       test_size=0.3,
                                       random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
knn_model = KNeighborsClassifier()
log_model = LogisticRegression()
dt_model = DecisionTreeClassifier()
```

```python
from sklearn.metrics import accuracy_score
```

```python
def model_training():
    knn_model.fit(X, Y)
    knn_predictions = knn_model.predict(x_test)
    knn_score = accuracy_score(knn_predictions, y_test)
    log_model.fit(X, Y)
    log_predictions = log_model.predict(x_test)
    log_score = accuracy_score(log_predictions, y_test)
    dt_model.fit(X, Y)
    dt_predictions = dt_model.predict(x_test)
    dt_score = accuracy_score(dt_predictions, y_test)
    
    print("KNN accuracy - ", knn_score)
    print("Log accuracy - ", log_score)
    print("Decision accuracy - ", dt_score)
```

```python
model_training()
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear')
```

```python
svm_model.fit(X, Y)
```

```python
svm_predictions = svm_model.predict(x_test)
```

```python
svm_score = accuracy_score(svm_predictions, y_test)
```
