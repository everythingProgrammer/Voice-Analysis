```python
import pandas as pd
```

```python
data = pd.read_csv("Voice-Dataset.csv")
```

```python
data.head(3)
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
data['label']= label_encoder.fit_transform(data['label'])
```

```python
data.head()
```

```python
data.columns
```

```python
columns = data.columns
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_data = data[columns[0:20]]
training_data.head()
```

```python
output_data = data['label']
output_data.head()
```

##### `neighbors = 5`  accuracy score = 0.7050
#####  ` neighbors = 10 ` accuracy score = 0.7018
##### `neighbor = 15 ` accuracy score = 0.7018

```python
from sklearn.model_selection import train_test_split
```

```python
X, x_test, Y, y_test = train_test_split(training_data, output_data, test_size=0.2, random_state = 42)
```

```python
model = KNeighborsClassifier(n_neighbors=15)
```

```python
model.fit(X, Y)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test, prediction)
score
```
